```python
import pandas as pd
import numpy as np
```

```python
school_means = {'Sleep': 7.9,
 'Personal Care': 1,
 'Household Activities': .6,
 'Work': .5,
 'Education': 7,
 'Shopping': .5,
 'Eating/Drinking': 1.5,
 'Leisure': 2,
 'Athletics': 2,
 'Other': 1}
```

```python
break_means = {'Sleep': 8.5,
 'Personal Care': 1.5,
 'Household Activities': 1,
 'Work': .5,
 'Education': .6,
 'Shopping': 1.4,
 'Eating/Drinking': 2,
 'Leisure': 5,
 'Athletics': 2.5,
 'Other': 1}
```

```python
work_means = {'Sleep': 7.9,
 'Personal Care': 1,
 'Household Activities': .6,
 'Work': 8,
 'Education': .5,
 'Shopping': .5,
 'Eating/Drinking': 1.5,
 'Leisure': 2,
 'Athletics': 1,
 'Other': 1}
```

```python
def gen_day_dat(month, day):
    if month == 12:
        schedule = break_means
    if month in [6,7,8]:
        if 30 * month + day % 7 <= 1:
            schedule = break_means
        else:
            schedule = work_means
    else:
        schedule = school_means
    
    day_dat = []
    total = 0
    for key in schedule:
        day_dat.append(np.random.uniform(schedule[key] * .8, schedule[key] * 1.2))
    day_dat = np.array(day_dat)
    day_dat = day_dat / np.sum(day_dat) * 24
    
    day_dic = {}
    for i in range(len(schedule)):
        day_dic[list(schedule.keys())[i]] = day_dat[i]
    
    return day_dic
```

```python
daily_activity = {}

# For simplicity, I'm just saying 30 days/month
for month in range(1, 13):
    for day in range(1, 31):
        day_dat = gen_day_dat(month, day)
        daily_activity[(month, day)] = day_dat
```

```python
activity_df = pd.DataFrame.from_dict(daily_activity, orient = "index")
```
